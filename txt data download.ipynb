```python
import pickle
with open("cik.txt", "rb") as fp:   # Unpickling
    cik_list = pickle.load(fp)
```

```python
import pickle
with open("ticker.txt", "rb") as fp:   # Unpickling
    ticker_list = pickle.load(fp)
```

```python
cik_list[:2]
```

```python
downloaded=[]
```

```python
remaining=[x for x in cik_list if x not in downloaded]
for ciks in remaining:
    url='https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK='+ciks+'&type=10-k&dateb=&owner=include&count=40'

    ticker=ticker_list[cik_list.index(ciks)]
    
    soup=BeautifulSoup(requests.get(url = url).content,'html.parser')
    if len(soup.find_all('table'))>2:
        ten_k_list_table=soup.find_all('table')[2]
        for j in ten_k_list_table.find_all('tr'):
            cols_j=j.find_all('td')
            if len(cols_j)!=0:
                date=cols_j[3].get_text()
                url_2='https://www.sec.gov/'+(cols_j[1].find('a', {'href':True, 'id':'documentsbutton'}))['href']
                soup=BeautifulSoup(requests.get(url = url_2).content,'html.parser')
                
                doc_list_table=soup.find_all('table')[0]
                
                if ('10-K' in doc_list_table.find_all('tr')[1].find_all('td')[3].get_text()):
                    txt_url=doc_list_table.find_all('tr')[1].find_all('td')[2].find('a')['href']
                    if ('.htm' in txt_url):
                        with open(ticker+"@"+date +'.htm', "wb") as f:
                            f.write(requests.get('https://www.sec.gov/'+txt_url).content)
                    elif ('.txt' in txt_url):
                        with open(ticker+"@"+date +'.txt', "wb") as f:
                            f.write(requests.get('https://www.sec.gov/'+txt_url).content)
                    else:
                        txt_url=doc_list_table.find_all('tr')[-1].find_all('td')[2].find('a')['href']
                        with open(ticker+"@"+date +'.txt', "wb") as f:
                            f.write(requests.get('https://www.sec.gov/'+txt_url).content)
                    sleep(1)
    downloaded.append(ciks)
```
